## Dosyayı tanımlama için gerekli forma getirmek

```python
blur = cv2.GaussianBlur(frame, (5,5),0)
```

```python
plt.imshow(frame)
```

```python
hls = cv2.cvtColor(frame, cv2.COLOR_BGR2HLS)
```

```python
plt.imshow(hls)
# hue lightness saturation
```

```python
ret, th = cv2.threshold(hls, 127, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(th)
```

```python
    lower = np.array([0,102,51])
    upper = np.array([255,255,255])
```

```python
mask = cv2.inRange(th, lower, upper)
```

```python
plt.imshow(mask)
```

```python
edges = cv2.Canny(mask, 75, 150)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
```

```python
plt.imshow(edges)
```

```python
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 0, maxLineGap=50)
```

```python
if lines is not None:
        for line in lines:
            x, y, w, h = line[0]
            cv2.line(blur, (x,y),(w,h), (0,255,0), 4)
            
```

```python
plt.imshow(blur)
```
